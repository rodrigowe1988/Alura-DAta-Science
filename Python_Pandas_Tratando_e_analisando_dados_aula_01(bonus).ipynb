# Trabalhando com base de dados de formatos diferentes

```python
import pandas as pd
```

## .JSON

```python
#trabalhando com arquivos do tipo json
json = open('/content/aluguel.json')
print(json.read())
```

```python
#para importar como dataframe do pandas
df_json = pd.read_json('aluguel.json')
df_json.head()
```

## .TXT

```python
#trabalhando com arquivos no formato txt
txt = open('aluguel.txt')
print(txt.read())
```

```python
#para importar como dataframe do pandas
df_txt = pd.read_table('aluguel.txt')
df_txt
```
